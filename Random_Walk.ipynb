```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def random_walk(n_steps=1000):
    steps = np.random.choice([-1, 1], size=n_steps)
    position = np.cumsum(steps)
    return position
```

```python
n_walks = 10
n_steps = 1000

walks = [random_walk(n_steps) for _ in range(n_walks)]
```

```python
plt.figure(figsize=(12, 6))

for walk in walks:
    plt.plot(walk, alpha=0.7)

plt.title("Simulation de marches aléatoires (1D)")
plt.xlabel("Étapes")
plt.ylabel("Position")
plt.grid(True)
plt.show()
```

```python
final_positions = [walk[-1] for walk in walks]

print("Positions finales :", final_positions)
print("Moyenne :", np.mean(final_positions))
print("Variance :", np.var(final_positions))
```
